```python
import pandas as pd 
import folium
import json
```

```python
df = pd.read_csv("data/home_value.csv")

df['zipcode'] = df['zipcode'].astype('int').astype('str')
df
```

```python
orlando_geo = r'data/city_orlando.geojson'
m = folium.Map(location=[28.5384, -81.3789], zoom_start=11)
choropleth = folium.Choropleth(
    geo_data = orlando_geo,
    name="choropleth",
    data = df,
    key_on = 'feature.properties.name',
    columns = ['zipcode','price'],
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity = 0.3,
    legend_name = "Zillow Home Value",
    highlight = True).add_to(m)

for i in range(0, len(df)):
    iframe = folium.IFrame('Zipcode: ' + df.iloc[i]['zipcode'] + '<br>' + 'Price: $' + str(df.iloc[i]['price']))
    popup = folium.Popup(iframe, min_width=140, max_width=100)
    folium.Marker(
        location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
        icon=folium.Icon(color='darkblue'),
        popup=popup
    ).add_to(m)

folium.LayerControl().add_to(m)

m
```
